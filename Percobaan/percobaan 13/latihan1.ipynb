```python
import pandas as pd
from collections import Counter

# Path ke file CSV
file_path = r"J:\Drive Saya\analisa sentimen git\percobaan 13\data_sentimen.csv"

# Membaca file CSV
data = pd.read_csv(file_path)

# Fungsi untuk menghitung frekuensi kata pada kolom tertentu
def get_top_words(data_column, top_n=20):
    words = []
    for tokens in data_column:
        try:
            words += eval(tokens)  # Mengubah string list kembali menjadi list
        except:
            pass
    return Counter(words).most_common(top_n)

# Fungsi untuk mendapatkan contoh kalimat
def get_example_sentences(word, data, column_tokens, column_content, n=10):
    examples = []
    for idx, row in data.iterrows():
        if word in eval(row[column_tokens]):
            examples.append(row[column_content])
        if len(examples) >= n:
            break
    return examples

# Hitung frekuensi kata secara keseluruhan
top_20_words = get_top_words(data['data_Preprocessing'], top_n=20)

# Filter data positif dan negatif
data_positive = data[data['sentimen'] == 'positive']
data_negative = data[data['sentimen'] == 'negative']

# Hitung kata paling sering muncul untuk sentimen positif dan negatif
top_positive_words = get_top_words(data_positive['data_Preprocessing'], top_n=5)
top_negative_words = get_top_words(data_negative['data_Preprocessing'], top_n=5)

# Output format yang diminta
print("Kata yang sering muncul:")
for word, count in top_20_words:
    print(f"{word} : {count}")

print("\nKata positif yang sering muncul (5):")
for word, count in top_positive_words:
    print(f"{word} : {count}")
    examples = get_example_sentences(word, data_positive, 'data_Preprocessing', 'content', n=10)
    print(f"Contoh kalimat {word} (10):")
    for sentence in examples:
        print(f"- {sentence}")

print("\nKata negatif yang sering muncul (5):")
for word, count in top_negative_words:
    print(f"{word} : {count}")
    examples = get_example_sentences(word, data_negative, 'data_Preprocessing', 'content', n=10)
    print(f"Contoh kalimat {word} (10):")
    for sentence in examples:
        print(f"- {sentence}")
```
