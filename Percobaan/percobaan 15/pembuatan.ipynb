```python
import csv
import ast

def tampilkan_driver_dengan_nomor(file_path):
    """
    Menampilkan nomor urut dan nilai dari atribut 'content' jika 'driver' ada di 'stemmed_tokens'
    dan 'score' berada di antara 1 dan 3.

    Args:
        file_path (str): Path menuju file CSV.
    """
    try:
        with open(file_path, 'r', encoding='utf-8') as csvfile:
            reader = csv.DictReader(csvfile)
            nomor_urut = 1
            for row in reader:
                try:
                    stemmed_tokens = ast.literal_eval(row['stemmed_tokens'])
                    score = int(row['score'])
                    if 'driver' in stemmed_tokens and 1 <= score <= 3:
                        print(f"{nomor_urut}. Content: {row['content']}")
                        nomor_urut += 1
                except (ValueError, SyntaxError) as e:
                    print(f"Error pada baris: {nomor_urut}, Data: {row}. Error: {e}")
                    nomor_urut += 1 # Tetap increment agar nomor urut tetap berlanjut
    except FileNotFoundError:
        print(f"Error: File tidak ditemukan di path: {file_path}")
    except Exception as e:
        print(f"Terjadi kesalahan: {e}")

# Ganti path file dengan path Anda
file_path = r'C:\Users\ghiff\Documents\kuliah\Analisa_Sentimen_Gojek\percobaa15\new-data.csv'
tampilkan_driver_dengan_nomor(file_path)
```
